```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('framingham.csv') 
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head(10)
```

```python
missing = df.isnull().sum()
print("Missing values:\n", missing)
```

```python
# Remove rows with missing values
df_cleaned = df.dropna()
```

```python
print("Missing values after cleaning:\n", df_cleaned.isnull().sum())
```

*Count plot of the target variable*

__Distribution of Key Features__

```python
# A count plot provides a clear view of how many individuals developed CVD vs. those who didn’t.
sns.countplot(x='TenYearCHD', data=df_cleaned, palette='Set2', hue='TenYearCHD')
plt.title('Target Variable Distribution (TenYearCHD)')
plt.xlabel('Heart Disease within 10 Years')
plt.ylabel('Count')
plt.legend([],[], frameon=False)  # Removes the legend if it's not needed
plt.show()
```

*Distribution of age*

```python

sns.histplot(df_cleaned['age'], kde=True, bins=20, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

*Distribution of cholesterol*

```python
sns.histplot(df_cleaned['totChol'], kde=True, bins=20, color='red')
plt.title('Cholesterol Distribution')
plt.xlabel('Total Cholesterol')
plt.ylabel('Frequency')
plt.show()
```

*Distribution of systolic blood pressure*

```python
sns.histplot(df_cleaned['sysBP'], kde=True, bins=20, color='green')
plt.title('Systolic Blood Pressure Distribution')
plt.xlabel('Systolic Blood Pressure')
plt.ylabel('Frequency')
plt.show()
```

Histograms with KDE plots help visualize the distribution of numerical features.

__Correlation Analysis__

*Correlation heatmap*

```python
correlation_matrix = df_cleaned.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

The heatmap reveals the strength of relationships between variables (e.g., high systolic BP and risk of heart disease).


__Feature Analysis by Target Variable__

*Age distribution by target variable*


```python
sns.boxplot(x='TenYearCHD', y='age', data=df_cleaned, palette='Set3', hue='TenYearCHD')
plt.title('Age vs. TenYearCHD')
plt.xlabel('Heart Disease within 10 Years')
plt.ylabel('Age')
plt.legend([],[], frameon=False)  
plt.show()
```

*Cholesterol distribution by target variable*

```python
 
sns.boxplot(x='TenYearCHD', y='totChol', data=df_cleaned, palette='Set2', hue='TenYearCHD')
plt.title('Total Cholesterol vs. TenYearCHD')
plt.xlabel('Heart Disease within 10 Years')
plt.ylabel('Total Cholesterol')
plt.legend([],[], frameon=False) 
plt.show()
```

*Systolic Blood Pressure by target variable*

```python

sns.boxplot(x='TenYearCHD', y='sysBP', data=df_cleaned, palette='Set1', hue='TenYearCHD')
plt.title('Systolic Blood Pressure vs. TenYearCHD')
plt.xlabel('Heart Disease within 10 Years')
plt.ylabel('Systolic Blood Pressure')
plt.legend([],[], frameon=False) 
plt.show()
```

Box plots compare distributions of numerical features between those who developed CVD and those who didn’t

__Analyzing Categorical Features__

*Gender and TenYearCHD*

```python
sns.countplot(x='male', hue='TenYearCHD', data=df_cleaned, palette='muted')
plt.title('Gender vs. TenYearCHD')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.legend(title='Heart Disease (10 Years)', loc='upper right')
plt.show()
```

*Smoking and TenYearCHD*

```python
sns.countplot(x='currentSmoker', hue='TenYearCHD', data=df_cleaned, palette='cool')
plt.title('Smoking vs. TenYearCHD')
plt.xlabel('Current Smoker (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.legend(title='Heart Disease (10 Years)', loc='upper right')
plt.show()
```

Count plots with hue analyze categorical variables like gender and smoking status.

__Insights from the analysis__


Target Variable:

*The dataset is imbalanced, with fewer individuals developing heart disease within 10 years.*

Key Risk Factors:

*Age, cholesterol levels, and systolic blood pressure show strong correlations with heart disease risk.*
*Smokers and males appear more likely to develop heart disease.*

Feature Comparisons:

*Higher systolic BP and cholesterol levels are observed in individuals with heart disease.*
